## Introducción

La regresión lineal es el algoritmo de aprendizaje supervisado más popular. Se utiliza para predecir el valor de una variable en función de otras que, el investigador supone, están relacionadas con ella. Por ejemplo, puede emplearse para pronosticar las ventas en los próximos meses en función de los gastos de publicidad.

En detalle, la regresión lineal es un tipo de análisis que plantea una relación lineal entre dos o más variables aleatorias: una variable respuesta (dependiente) y una variable explicativa (independiente) en el caso del modelo de regresión simple, o una variable respuesta y múltiples variables explicativas en el caso del modelo de regresión múltiple. La ecuación que relaciona estas variables es la siguiente:

$$
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_pX_p + \epsilon,
$$

donde $\beta_0, ..., \beta_p$ son los coeficientes, $X_1, ..., X_p$ las caracteristicas o variables explicativas, $Y$ es la variable respuesta y $\epsilon$ es un término aleatorio o error que se supone se distribuye normal con media igual a cero y varianza constante: $\epsilon \sim \mathcal{N}(0,\sigma^2)$. Por lo tanto, los posibles valores que tome la variable $Y$ para ciertos valores de las variables explicativas se distribuyen alrededor de una recta verdadera o recta de regresión. El objetivo es hallar esa recta de regresión:

$$
E[Y|\mathbb{X}] = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_pX_p
$$

Sin embargo, a este nivel se está tratando con variables aleatorias gobernadas por distribuciones de probabilidad desconocidas por el investigador, y de las que solo puede extraer información parcial mediante la observación. En la práctica, la verdadera recta de regresión se aproxima mediante el planteamiento de un modelo de regresión que utiliza los valores observados de las variables aleatorias que se supone se relacionan de forma lineal. 

## Historia

El primer registro que hasta la fecha se conoce sobre el método de regresión lineal es la descripción realizada por Legendré en 1805. Por su parte, Gauss (1821)[$^1$](#Referencias) desarrolló con mayor profundidad el método de mínimos cuadrados ordinarios[$^2$](#Referencias). El término regresión lo utilizó por primera vez el médico Francis Galton, primo de Charles Darwin, en su libro Natural Inheritance (1889). Este trabajo se centró en la descripción de los rasgos físicos de los descendientes a partir de los de sus padres. Galton estudió la altura de padres e hijos en mil registros de grupos familiares y concluyó que padres muy altos tenían hijos que también eran altos pero no tanto como los primeros, es decir, los hijos heredaban parte de esa altura pero en general tendían a la media. El autor generalizó esta tendencia con la idea de "regresión a la media"[$^3$](#Referencias)

## Algoritmo

### Regresión Lineal Simple

Sea $Y$ la variable aleatoria respuesta y $X$ una variable aleatoria explicativa. El modelo de regresión lineal es:

$$
Y = \beta_0 + \beta_1X + \epsilon, \ \ \ \epsilon \sim \mathcal{N}(0,\sigma^2)
$$

El objetivo es hallar la recta de regresión estimando los coeficientes $\beta_0$ y $\beta_1$:

$$
E[Y|X] = \beta_0 + \beta_1X
$$

La Figura 1[$^4$](#Referencias) representa gráficamente la cuestión:

fig1.png

El problema es que el investigador solo puede observar $n$ realizaciones tanto de $Y$ como de $X$ cuando toma un muestra aleatoria de la población, es decir, cuando selecciona los datos que utilizará en su análisis. El modelo de regresión lineal para los datos observados es:

$$
y = b_0 + b_1x + e, \ \ \ e \sim \mathcal{N}(0,\sigma^2),
$$

y la recta de regresión que se ajusta a los datos observados se reduce a una ecuación lineal donde $b_0$ es el intercepto y $b_1$ la pendiente:

$$
E[y|X=x] = b_0 + b_1x
$$

En este caso, la Figura 2[$^5$](#Referencias) muestra la diferencia entre la recta de regresión verdadera y su estimación a partir de los datos:

fig2.png

A esta altura, interesa saber cómo calcular el intercepto y la pendiente de la ecuación lineal para obtener la estimación de la recta de regresión que "mejor" se ajuste a los datos observados. Para ello, se emplea el algoritmo de Mínimos Cuadrados Ordinarios que permite hallar los valores de los parámetros que minimizan la distancia entre $y$ y $E[y|X=x]$, es decir, que reducen al máximo el error de estimación (desviación vertical de la variable respuesta con respecto a su media, ver Figura 3[$^6$](#Referencias))

### Función de Pérdida

Sea el modelo de regresion lineal múltiple para la $i$-ésima observación de una muestra aleatoria de tamaño $n$:

$$
y_i = \beta_0 + \beta_1x_{1,i} + \beta_2x_{2,i} + ... + \beta_{p-1}x_{p-1,i} + e_i \\e_i \sim \mathcal{N}(0,\sigma^2), \ \ \ i=0,...,n-1
$$


Definiendo la matriz de características de tamaño $n\times p$ como: $\ \mathbb{X}=(\mathbb{x_0},\ \mathbb{x_1},...,\ \mathbb{x_{n-1}})^T\ $ con $\mathbb{x_i}=(1,\ x_{1,i},\ x_{2,i}, ...,\ x_{p-1,i})^T,\ $el vector de tamaño $n\times 1\ $para la variable respuesta: $\ \mathbb{Y} = (y_0,\ y_1,...,\ y_{n-1})^T,\ $ el vector de coeficientes de tamaño $p\times 1$: $\ \mathbb{\beta}=(\beta_0,\ \beta_1,...,\ \beta_{p-1})^T\ $ y el vector de errores de tamaño $n\times 1$ : $\ \mathbb{E}=(e_0,\ e_1,...,\ e_{p-1})^T,\ $el modelo en forma matricial está dado por:

$$
\mathbb{Y}=\mathbb{X}\mathbb{\beta}\ +\ \mathbb{E}
$$

Por su parte, el algoritmo de Minimos Cuadrados Ordinarios consiste en un problema de optimización: hallar el valor de los parámetros de la ecuación lineal que minimicen la función de pérdida conocida como Suma de los Cuadrados de los Errores (SSE por sus siglas en inglés). La función de pérdida es:

$$
SSE(\hat{\beta}) = \sum_{i=0}^{n-1}(y_i-\mathbb{x_i}^T\hat{\beta})^2
$$

En forma matricial la función de pérdida está dada por:

$$
SSE(\hat{\beta}) = (\mathbb{Y} - \mathbb{X}\hat{\beta})^T(\mathbb{Y} - \mathbb{X}\hat{\beta})
$$

El problema de optimización se reduce a resolver:

$$
\DeclareMathOperator*{\argmin}{argmin}
\hat{\beta} = \argmin_{\beta\ \in\ \mathbb{R}^p} SSE(\beta)
$$

De la ecuación $\ \partial(SSE)/\partial\beta=0\ $ se despeja el vector $\beta$. Bajo el supuesto de que $\ \mathbb{X}^T\mathbb{X}\ $ tiene rango completo por columna, la matriz Hessiana $\ \partial^2(SSE)/\partial\beta\partial\beta^T\ $es definida positiva, por tanto, en $\beta$ la función SSE alcanza un mínimo global. 

Entonces, el vector de coeficientes que minimiza la función de pérdida está dada por:

$$
\hat{\beta} = [\mathbb{X}^T\mathbb{X}]^{-1}\mathbb{X}^T\mathbb{Y}
$$

Por último, el vector de respuestas estimadas es:

$$
\mathbb{\hat{Y}} = \mathbb{X}\hat{\beta} = \mathbb{X}[\mathbb{X}^T\mathbb{X}]^{-1}\mathbb{X}^T\mathbb{Y}
$$

En la siguiente rutina se presentan las ideas planteadas hasta ahora y se implementa el algoritmo aplicando directamente las fórmulas.

```
# librerias
import numpy                  as np
import matplotlib.pyplot      as plt
%matplotlib inline
plt.style.use('ggplot')


np.random.seed(0)

# matriz de características y vector de respuestas
intercept = np.ones((100,1))
X = np.random.randint(0,100,100).reshape((100,1))
X = np.concatenate((intercept, X), axis=1)
y = X[:,1] + np.random.normal(loc=0, scale=15, size=100)

# vector de coeficientes estimados, vector de respuestas estimadas y SSE
beta = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
y_hat = np.dot(X, beta)
SSE = np.sum((y-y_hat)**2)
print(f'Parámetros estimados: {beta}', f'SSE: {SSE}', sep='\n')

# y vs X, recta de regresión estimada
plt.scatter(X[:,1], y, c='r', s=3)
plt.plot(X[:,1], y_hat, c='b', label='Recta de regresión')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()
```

## Aplicación práctica

En el ejercicio siguiente se ajusta un modelo de regresión lineal a una muestra de viviendas localizadas en la ciudad de Boston, Estados Unidos. El objetivo es modelar la variable respuesta como una función lineal de ciertas características (variables predictoras) que, en principio, se consideran relevantes para explicar su variación. Para ello se estiman sus coeficientes (parámetros) en el modelo lineal mediante el algoritmo de Mínimos Cuadrados Ordinarios.

La variable respuesta `MEDV` es el precio mediano (miles de dólares) de las viviendas. Las características son el número promedio de habitaciones `RM`, el porcentaje de la población de estrato bajo que reside en el vecindario `LSTAT`, el procentaje de unidades habitacionales de la vivienda construídas antes de 1940 `AGE` y la concentración de óxido nítrico en el aire (en partes por 10 millones) `NOX`.

Así, el modelo propuesto es el siguiente:

$$
MEDV = \beta_0 + \beta_1RM + \beta_2LSTAT + \beta_3Age + \beta_4NOX + \epsilon
$$

```
# librerias
import pandas                 as pd
from sklearn.datasets         import load_boston
from sklearn.model_selection  import train_test_split
from sklearn.linear_model     import LinearRegression
```

```
# carga de los datos en memoria
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['MEDV'] = boston_dataset.target 

boston.head()
```

```
# descripción del dataset
print(boston_dataset.DESCR)
```

```
# representación gráfica de los datos: histogramas y diagramas de dispersión de las variables incluídas en el modelo
pd.plotting.scatter_matrix(boston[['MEDV', 'RM', 'LSTAT', 'AGE', 'NOX']], alpha=0.5, figsize=(15, 15))
plt.tight_layout()
```

```
# variable respuesta y matriz de caracteristicas
X = boston[['RM', 'LSTAT', 'AGE', 'NOX']]
y = boston_dataset.target

# conjunto de datos de entrenamiento y validación
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```
# # ajuste del modelo a los datos de entrenamiento
ols = LinearRegression()
ols.fit(X_train, y_train)

# parámetros estimados
coef = np.append(ols.intercept_, ols.coef_)
print(f'Parámetros estimados: {coef}')
```

El modelo estimado para el precio mediano de la vivienda en la ciudad de Boston es el siguiente:

$$
MEDV = -0.3959 + 5.2480RM + -0.6273LSTAT + 0.0166Age + -5.9530NOX
$$

```
# predicción e indicadores de desempeño: RMSE y Coeficiente de determinación R^2
y_pred = ols.predict(X_test)
resids = y_test - y_pred

rmse = np.sqrt(np.mean(resids**2))
r2 = ols.score(X_train, y_train)
print(f'RMSE: {rmse}', f'Coeficiente de determinación: {r2}', sep='\n')
```

```
# análisis gráfico: respuesta ajustada vs respuesta observada, normal qq-plot de los residuos 
from scipy.stats import probplot
fig, ax = plt.subplots(1,2, figsize=(15,5))

ax[0].scatter(y_test, y_pred, s=3)
ax[0].set_title('Precio mediano estimado vs Precio mediano observado (miles de dólares)')
ax[0].set_xlabel('Precio mediano observado de la vivienda')
ax[0].set_ylabel('Precio mediano estimado de la vivienda')
probplot(resids, dist='norm', plot=ax[1])

fig.tight_layout()
```

El precio mediano estimado de una vivienda en la ciudad de Boston con las siguientes caracteristicas: 4 habitaciones, la población de estrato bajo en el vecindario corresponde al 2% del total, el porcentaje de las unidades habitacionales que son antiguas es 20% y la concentración de óxido nítrico (partes por 10 millones) es 0.4:

```
datapoint = np.array([[5, 2, 20, 0.4]])
pred = round(ols.predict(datapoint)[0]*1000, 2)
print(f'El precio mediano estimado de la vivienda es {pred} dólares.')
```
