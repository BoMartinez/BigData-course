NYC 311
---------

[Download link](https://nycopendata.socrata.com/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9)

All 311 service requests since 2010 in New York City

```python
# descarga del dataset NYC311 en el directorio que corresponde
import os
import requests

path = os.path.join('data','nyc311')
if not os.path.exists(path):
  os.makedirs(path)

print('Downloading NYC311 dataset... ', end='', flush=True)
r = requests.get('https://data.cityofnewyork.us/resource/erm2-nwe9.json')
with open(os.path.join(path, 'nyc311.json'), 'wb') as f:
  f.write(r.content)
print('Done!', flush=True)
```
